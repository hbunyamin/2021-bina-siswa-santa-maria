```python
import sys
 
 
# Function to find the length of the longest increasing subsequence
def LIS(A, i, n, prev):
 
    # Base case: nothing is remaining
    if i == n:
        return 0
 
    # case 1: exclude the current element and process the
    # remaining elements
    excl = LIS(A, i + 1, n, prev)
    print(excl)
 
    # case 2: include the current element if it is greater
    # than the previous element in LIS
    incl = 0
    if A[i] > prev:
        incl = 1 + LIS(A, i + 1, n, A[i])
 
    # return the maximum of the above two choices
    return max(incl, excl)
```

```python
A = [2, 14, 7, 20, 5, 3, 8, 11, 18, 4, 10, 12, 1, 6, 9, 19, 15, 16, 13, 17]
```

```python
print("The length of the LIS is", LIS(A, 0, len(A), -sys.maxsize))
```
