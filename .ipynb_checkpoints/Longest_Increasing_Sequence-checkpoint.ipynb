## Counting the Valleys

```python
steps = 8
path = "DDUUUUDD"
```

```python
def countingValleys(steps, path):
    num_valleys = 0
    prev_level = 0
    curr_level = 0
    for step in range(steps):
        prev_level = curr_level
        if path[step] == "U":
            curr_level += 1
        else:
            curr_level -= 1
        if prev_level < 0 and curr_level == 0:
            num_valleys+=1
    return num_valleys        
```

```python
countingValleys(8, path)
```

## Jumping the Clouds

```python
c1 = [0,1,0,0,0,1,0]
c2 = [0,0,1,0,0,1,0]
c3 = [0,0,0,0,1,0]
```

```python
def jumpingOnClouds(c):
    curr_cloud = 0
    num_of_jumps = 1
    next_indeks = 1
    while next_indeks < len(c):
        if c[next_indeks] == 0  and (next_indeks+1) < len(c) and c[next_indeks+1] == 0:
            curr_cloud = c[next_indeks+1]
            next_indeks+=2
            num_of_jumps += 1
        elif c[next_indeks] == 0  and (next_indeks+1) < len(c):
            curr_cloud = c[next_indeks]
            next_indeks+=1
            num_of_jumps += 1
        else:
            curr_cloud =c[next_indeks]
            next_indeks+=2
            num_of_jumps += 1
    return num_of_jumps
```

```python
jumpingOnClouds(c1), jumpingOnClouds(c2), jumpingOnClouds(c3)
```

```python
import sys
 
 
# Function to find the length of the longest increasing subsequence
def LIS(A, i, n, prev):
 
    # Base case: nothing is remaining
    if i == n:
        return 0
 
    # case 1: exclude the current element and process the
    # remaining elements
    excl = LIS(A, i + 1, n, prev)
    print(excl)
 
    # case 2: include the current element if it is greater
    # than the previous element in LIS
    incl = 0
    if A[i] > prev:
        incl = 1 + LIS(A, i + 1, n, A[i])
 
    # return the maximum of the above two choices
    return max(incl, excl)
```

```python
A = [2, 14, 7, 20, 5, 3, 8, 11, 18, 4, 10, 12, 1, 6, 9, 19, 15, 16, 13, 17]
```

```python
print("The length of the LIS is", LIS(A, 0, len(A), -sys.maxsize))
```
